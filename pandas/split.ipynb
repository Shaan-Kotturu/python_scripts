```python
import pandas as pd

df = pd.read_csv("dim_hotels.csv")
#df.head()
#df['city'].unique()

#for city in df['city'].unique():
#    print(city)

writer = pd.ExcelWriter("dim_hotels_by_city.xlsx", engine='xlsxwriter')
for city in df['city'].unique():
    city_df = df[df['city'] == city]
    city_df.to_excel(writer, sheet_name=city[:31], index=False)  # Sheet names max length is 31 characters

writer.close()
```
